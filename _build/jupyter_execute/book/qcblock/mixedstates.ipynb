–≤–µ–∫—Ç–æ—Ä–æ–≤ –æ–ø–µ—Ä–∞—Ç–æ—Ä–∞ –Ω–∞–±–ª—é–¥–∞–µ–º–æ–π. –ê –º–æ–∂–µ–º –ª–∏ –º—ã –ø–æ—Å–º–æ—Ç—Ä–µ—Ç—å –Ω–∞ —ç—Ç–æ —Å –¥—Ä—É–≥–æ–π —Å—Ç–æ—Ä–æ–Ω—ã: –∫–∞–∫ –ø—Ä–æ–µ–∫—Ü–∏—é –æ–ø–µ—Ä–∞—Ç–æ—Ä–∞ –Ω–∞–±–ª—é–¥–∞–µ–º–æ–π 
–Ω–∞ —Å–æ—Å—Ç–æ—è–Ω–∏–µ —Å–∏—Å—Ç–µ–º—ã?

–û—Ç–≤–µ—Ç –¥–æ–≤–æ–ª—å–Ω–æ –ø—Ä–æ—Å—Ç–æ–π: –º–æ–∂–Ω–æ –∑–∞–ø–∏—Å–∞—Ç—å —Å–æ—Å—Ç–æ—è–Ω–∏–µ —Å–∏—Å—Ç–µ–º—ã $\ket{\phi}$ –≤ –≤–∏–¥–µ _–æ–ø–µ—Ä–∞—Ç–æ—Ä–∞-–ø—Ä–æ–µ–∫—Ç–æ—Ä–∞_

$$

$$ (eqn:rho_pure)

(–∏–Ω—ã–º–∏ —Å–ª–æ–≤–∞–º–∏, –º—ã –∏—Å–ø–æ–ª—å–∑—É–µ–º –∫—Ä–æ–Ω–µ–∫–µ—Ä–æ–≤—Å–∫–æ–µ –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏–µ –≤–µ–∫—Ç–æ—Ä–∞ —Å–æ—Å—Ç–æ—è–Ω–∏—è –Ω–∞ —Å–∞–º–æ–≥–æ —Å–µ–±—è, —á—Ç–æ–±—ã –ø–æ–ª—É—á–∏—Ç—å –º–∞—Ç—Ä–∏—Ü—É

–ò—Å–ø–æ–ª—å–∑—É—è –æ–ø–µ—Ä–∞—Ç–æ—Ä-–ø—Ä–æ–µ–∫—Ç–æ—Ä, –º–æ–∂–Ω–æ –∑–∞–ø–∏—Å–∞—Ç—å –æ–∂–∏–¥–∞–µ–º–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –∏–∑–º–µ—Ä–µ–Ω–∏–µ –≤–µ–ª–∏—á–∏–Ω—ã, –æ–ø–∏—Å—ã–≤–∞–µ–º–æ–µ –æ–ø–µ—Ä–∞—Ç–æ—Ä–æ–º $A$ –∫–∞–∫

$$
\braket{A} = Tr(A \rho) = Tr(\rho A)
$$

—Ä–∞—Å—Å–º–æ—Ç—Ä–∏–º —Å–∏—Ç—É–∞—Ü–∏—é, –∫–æ–≥–¥–∞ –æ–ø–∏—Å—ã–≤–∞–µ–º—É—é –∫–≤–∞–Ω—Ç–æ–≤—É—é —Å–∏—Å—Ç–µ–º—É $\Phi$ –º–æ–∂–Ω–æ —Ä–∞–∑–¥–µ–ª–∏—Ç—å –Ω–∞ –¥–≤–µ –ø–æ–¥—Å–∏—Å—Ç–µ–º—ã, $\phi$ –∏ $\psi$, 
–∏ —Å–æ—Å—Ç–æ—è–Ω–∏–µ —ç—Ç–æ–π —Å–∏—Å—Ç–µ–º—ã $\ket{\Phi}$ —è–≤–ª—è–µ—Ç—Å—è —Å—É–ø–µ—Ä–ø–æ–∑–∏—Ü–∏–µ–π —Å–æ—Å—Ç–æ—è–Ω–∏–π –¥–≤—É—Ö –ø–æ–¥—Å–∏—Å—Ç–µ–º:

CNOT –∏–ª–∏ CZ, –∫–æ—Ç–æ—Ä—ã–π –∏—Å–ø–æ–ª—å–∑—É–µ—Ç—Å—è –≤ –ª–µ–∫—Ü–∏–∏ –ø—Ä–æ [–ì—Ä–∞–¥–∏–µ–Ω—Ç—ã –∫–≤–∞–Ω—Ç–æ–≤—ã—Ö —Å—Ö–µ–º](gradients). 

–í –∫–∞—á–µ—Å—Ç–≤–µ –ø—Ä–∏–º–µ—Ä–∞ –¥–∞–≤–∞–π—Ç–µ –ø–æ—Å–º–æ—Ç—Ä–∏–º, –∫–∞–∫ –º–æ–∂–Ω–æ —Å–æ–∑–¥–∞—Ç—å –∑–∞–ø—É—Ç–∞–Ω–Ω–æ–µ —Å–æ—Å—Ç–æ—è–Ω–∏–µ –≤ PennyLane.

–ù–∞—á–Ω–µ–º —Å –∏–º–ø–æ—Ä—Ç–æ–≤ –∏ —Å–æ–∑–¥–∞–Ω–∏—è –¥–≤—É—Ö–∫—É–±–∏—Ç–Ω–æ–π —Å—Ö–µ–º—ã:

````python
import pennylane as qml
from pennylane import numpy as np

dev = qml.device("default.qubit", wires=2)
````

–î–∞–ª–µ–µ –ø—Ä–∏–º–µ–Ω–∏–º –∫ –ø–µ—Ä–≤–æ–º—É –∫—É–±–∏—Ç—É –æ–ø–µ—Ä–∞—Ü–∏—é –ø–æ–≤–æ—Ä–æ—Ç–∞ $\hat{RX}$, –∑–∞–ø—É—Ç–∞–µ–º –∫—É–±–∏—Ç—ã —Å –ø–æ–º–æ—â—å—é CNOT –∏ –¥–∞–ª–µ–µ –æ—Ü–µ–Ω–∏–º 
–∑–∞–ø—É—Ç–∞–Ω–Ω–æ—Å—Ç—å —Å –ø–æ–º–æ—â—å—é –∏–∑–º–µ—Ä–µ–Ω–∏—è –æ–ø–µ—Ä–∞—Ç–æ—Ä–∞ –ü–∞—É–ª–∏ $\hat{\sigma^z}$:

````python
@qml.qnode(dev)
def circuit(param):
    qml.RX(param, wires=0)
    qml.CNOT(wires=[0, 1])
    return qml.expval(qml.PauliZ(0)), qml.expval(qml.PauliZ(1))
````

–í —ç—Ç–æ–º –ø—Ä–∏–º–µ—Ä–µ –∑–Ω–∞—á–µ–Ω–∏–µ –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π `param` –æ–ø—Ä–µ–¥–µ–ª—è–µ—Ç —Å—Ç–µ–ø–µ–Ω—å –∑–∞–ø—É—Ç–∞–Ω–Ω–æ—Å—Ç–∏, –∏ –¥–ª—è $\pi/2$ –∑–∞–ø—É—Ç–∞–Ω–Ω–æ—Å—Ç–∏ –±—É–¥–µ—Ç 
–º–∞–∫—Å–∏–º–∞–ª—å–Ω–∞, –≤ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–µ –æ–±–∞ –∫—É–±–∏—Ç–∞ –±—É–¥—É—Ç –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ —Å–º–µ—à–∞–Ω–Ω—ã–º–∏ –∏ —Å—Ä–µ–¥–Ω–∏–π —Ä–µ–∑—É–ª—å—Ç–∞—Ç –∏–∑–º–µ—Ä–µ–Ω–∏—è –±—É–¥–µ—Ç –Ω—É–ª–µ–≤—ã–º:

````python
print(circuit(np.pi / 2))
````
